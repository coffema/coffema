````python
uni = df['Education'].unique()
uni2 = df['City'].unique()
uni3 = df['Gender'].unique()
uni4 = df['EverBenched'].unique()
````

````python
E= {}
for index ,feat in enumerate(uni):
  E[feat] = index

C= {}
for index ,feat in enumerate(uni2):
  C[feat] = index

G= {}
for index ,feat in enumerate(uni3):
  G[feat] = index

EB= {}
for index ,feat in enumerate(uni4):
  EB[feat] = index
````

````python
df['Education'].replace(E,inplace = True)
df['City'].replace(C,inplace = True)
df['Gender'].replace(G,inplace = True)
df['EverBenched'].replace(EB,inplace = True)
df
````

````python
n=pd.cut(df['Education'],bins=3)
n2=pd.cut(df['City'],bins=3)
n3=pd.cut(df['Gender'],bins=2)

n5=pd.cut(df['JoiningYear'],bins=8)
n6=pd.cut(df['PaymentTier'],bins=3)
n7=pd.cut(df['ExperienceInCurrentDomain'],bins=10)
n4=pd.cut(df['EverBenched'],bins=2)
n8=pd.cut(df['Age'],bins=30)
n9=pd.cut(df['LeaveOrNot'],bins=2)
n.value_counts()
````

````python
# Show the Plots for the following columns 
# Pie chart for the Education, JoiningYear, City, PaymentTier and Gender
# Bar chart for the ExperienceInCurrentDomain and EverBenched
# Histogram for the Age and LeaveOrNot
fig, axes = plt.subplots(nrows=3, ncols=4,figsize=(10,10))
n.value_counts().plot(kind='pie',ax=axes[0,0])
n5.value_counts().plot(kind='pie',ax=axes[0,1])
n2.value_counts().plot(kind='pie',ax=axes[0,2])
n6.value_counts().plot(kind='pie',ax=axes[0,3])
n3.value_counts().plot(kind='pie',ax=axes[1,0])
n7.value_counts().plot(kind='bar',ax=axes[1,1])
n4.value_counts().plot(kind='bar',ax=axes[1,2])
n8.value_counts().plot.hist(grid=True,ax=axes[2,1])
n9.value_counts().plot.hist(grid=True,ax=axes[2,2])
# code start here
  
# code end here
````
